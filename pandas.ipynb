# nettoyer des  données

Les données ne sont pas toujours de bonnes qualités avec parfois des valeurs manquantes, il est donc impératif de les traiter avant de les utiliser.

## supprimer les colonnes inutiles

Un jeu de donnée peut contenir des caractères parfois
inutiles, dans ce cas on utilises la fonctions drop

````python
data = data.drop(['Patient ID','Hematocrit'], axis = 1)
#axis = 1: signifie suivant l'axe des colonnes, quant à l'axe = 0 c'est  suivant les lignes
````

````python
# si on vérifie notre dataset nous remarquons la disparution de ces 2 caractères et la  réduction du nombre de colonnes à 109
data.head(10)
````

## supprimer les lignes contenant des valeurs manquantes

 Pour supprimer les lignes contenant des nan nous utiliserons la fontion dropna()

 **Remarque** : Par défaut, la méthode dropna() renvoie un nouveau DataFrame et ne modifiera pas l'original, mais vous pouvez le faire en y ajoutant en argument 

```
inplace = True 
```



````python
new_df = data.dropna()

print(new_df) 
````

## remplacer les nan

on utilise la fonction fillna()

````python
data.fillna(2)
#remplacer les valeur manquantes par une valeur quelconque, dans notre cas 2
#Nan : il fait mention de données manquantes, il est différent d'une valeur nulle
#la suppression des colonnes qui en contiennent peuvent fausser la véracité des résultats
````

````python
data["Hemoglobin"].fillna(5, inplace=True) 
#permet de remplace avec plus de précision les nan de la colonne hemoglobin par 5
print(data)
````

````python
#on peut aussi remplacer les nan par la moyenne, la médiane, le mode de la colonne spécifiée
x = data["Platelets"].mean()
data["Platelets"].fillna(x, inplace=True)
print(data["Platelets"])
````

**4**. **Correction des données**

Supposons que nous voulons remplacer à **l**'**indexe** **1** l'âge du patient de 17 ans par 15 ans, pour cela nous utiliserons la fonction loc(). Cette fonction permet donc de corriger avec exactitude la valeur d'une cellule.



````python
data.loc[1, 'Patient age quantile'] = 15
print(data['Patient age quantile'])
````

**Correlation**

La correlation permet d'indiquer le dégré de liaison entre deux caractéristiques. Plus elle est élévée on peut prédire le comportement de l'un en fonction de l'autre et inversement.


La correlation varie de -1 à 1.

exemple:

1 signifie qu'il existe une relation de 1 à 1 (une corrélation parfaite), et pour cet ensemble de données, chaque fois qu'une valeur augmente dans la première colonne, l'autre augmente également.

0,9(plus elle se rapproche de 1) est également une bonne relation, et si vous augmentez une valeur, l'autre augmentera probablement aussi.

-0,9(plus elle se rapproche de -1) serait une relation aussi bonne que 0,9, mais si vous augmentez une valeur, l'autre diminuera probablement.

0,2 signifie PAS une bonne relation, ce qui signifie que si une valeur augmente, cela ne signifie pas que l'autre augmentera.


# Dataframes et séries









## Notion de Dataframe et les séries

La bibliothèque pandas fai cas de la notion de *série* qui est un tableau numpy à une dimention à coté duquel on rajoute une colonne d'indexe.

L'ensemmble de séries forment un *dataframe*, donc un tableau à plusieurs dimentions.

## filtrer ou trier des données

Pandas permet d'éfectuer une suite d'opérations pour des résultats plus poussés.
Par exemple on désir savoir le nombre de passager de moins de 18 ans avec comme indexe le taux d'hémoglobine.

````python
data[data['Patient age quantile']<18]['Hemoglobin'].value_counts()
````

On peut encore faire ressortir la moyenne des différentes catégories d'âge en les groupant en plusieurs caractéristiques.


````python
data[data['Patient age quantile']<18].groupby(['Hemoglobin',]).mean()
````

La première ligne révèle qu'il y a 100% des patients **admis en soin régulier** qui ont **moins d'un an** dont l'hémoglobine a pour valeur"**-3.844355**".

Ici nous avons extrait une variable cible " âge", en particulier les mineurs,ensuite nous avons groupé chaque tranche d'âge selon le taux d'hémoglobine et enfin afficher la moyenne du tout en fonction des autres variables. 

````python
import pandas as pd #création d'une série
Dt = pd.Series(['April-10', 'April-11', 'April-12', 'April-13'])
print(Dt)
````

````python
#création d'une dataframe
dates=['April-10', 'April-11', 'April-12', 'April-13']
fruits=['Apple', 'Papaya', 'Banana', 'Mango']
prices=[3, 1, 2, 4]
df = pd.DataFrame({'Date':dates ,
                   'Fruit':fruits ,
                   'Price': prices})
sport = ['football', 'baseball', 'volleyball', 'basketball']
df1 = pd.DataFrame({'sport':sport})

#concaténation de 2 dataframes
result = pd.concat([df1, df], axis=1)
print(result) 
````

**Modifier des valeurs dans un** **dataframe**

Une fonction capable d'effectuer cette tâche es la fonction **replace()**

*   Remplacer une seule valeur


```
 df[column_name].replace([old_value], new_value)
```


*   Remplacer plusieurs valeurs par la même valeur


```
df[column_name].replace([old_value1, old_value2, old_value3], new_value)
```


*   Remplacer les valeurs multiples par des valeurs multiples


```
df[column_name].replace([old_value1, old_value2, old_value3],[new_value1, new_value2, new_value3])
```


*   Remplacer une valeur par une nouvelle valeur pour l’ensemble du DataFrame


```
df.replace([old_value], new_value)
```
