# I.les étapes basiques

## 1.Avant l'implémentation du modèle

 **1.1.visualiser et analyser les données**

 Loin d'être impérative cette étape est tout de même nécessaire pour comprendre les données et les relations qui les lient. On utilise pour cela les bibliothèques pandas, et seaborn ou matplolib pour visualiser les données.

**1.2.pré-traiter les données** 

dans un dataset les données peuvent contenir des chaines de caractère alors que la machine ne traite que les données numérique.
Il convient donc d'encoder nos données, et il existe plusieur types encodage.
Encodage ordinal
Il consiste à associer chaque catégorie ou classe d’une variable à une
valeur décimale unique.
On dispose des Transformers comme

* LabelEncodeur() : utilisé pour encoder les valeur de la colonnes Y
* OrdinalEncodeur() :utiliser pour encoder les valeurs de la colonnes X

Encodage one hot,
Chaque catégorie est représentée de manière binaire dans une colonne
qui lui est propre. On dispose aussi plusieurs Transformers :

* LabelBinarizer()

* MultiLabelBinarizer()

* OneHotEncoder

**1.3.La mise à l'échelle**

La plupart du temps, en machine Learning, les DataSet viennent avec des ordres de grandeurs différents. Cette différence d’échelle peut conduire à des performances moindres. Pour palier à cela, des traitements préparatoires sur les données existent. Notamment le Feature Scaling qui comprend la **Standardisation** et la **Normalisation**.
Cette étape n'est pas obligatoire pour le preprocessing, elle permet d'ameliorer la qualité des résultats dans certains cas.

**1.4.Segmenter les données**

On founi à la machine des données d'entrainement et des données de test.
Il ne faut jamais évaluer son modèle sur les mêmes données qui ont servi à l'entrainement. Pour cela on dicise le dataset généralement en données d'entrainement(80%) et données de test (20%). Cette segmentation peut se faire manuellement ou à l'aide de la fonction ``train_test_split()`` de sickitlearn.



## 2.Implémentation du modèle

 **2.1. Choisir un modèle**

La première chose à faire avant de faire du ML est de choisir le modèle approprié en fonction du problème dont il est question.

https://scikit-learn.org/stable/tutorial/machine_learning_map/index.html

**3.2. Paramétrer le modèle**

Après avoir choisi votre modèle il est important de préciser ses hyperparamètres

**4.3. Entrainer le modèle**

A la suite de cela, il faut entrainer le modèle sur les données X,Y (divisées en 2 tableaux Numpy) avec la methode **```fit(X,Y)```**.
La dimension de X est:[n_samples, n_features]
La dimension de Y est : [n_samples, 1]

**5.4. Evaluer le modèle**

Pour connaître la performance de notre modèle , on l'évalue avec la fonction **``score(X, Y)``** , lors de cette étape cette fonction utilise des données X pour faire des prédictions et ensuite les comparer aux variables Y présent dans la dataset. Une fois satisfait du score vous pouver effectuer de nouvelles prédictions avec **``predict(X)``**


## 3.Résumé

La structure basique de résolution d'un problème en ML se présente comme suit:

``model=LinearRegression(paramètres)``

``model = fit(X, Y)``

``model.score(X, Y)``

``moddel.predict(X)``

NB : D'autre modèles de ML que "LinearRegression" existe, comme le support vecteur machine, regression logistique, etc. Pour appliquer chacune d'elle , il faut juste remplacer le modèle souhaité en lieu et place de "LinearRegression".

# II.Ajout d'étapes amélioratives

## données de validation

Si l'on règle les paramètres de notre modèle sur les données du test_set la machine aura "indirectement" vu ces données, c'est pour cela qu'on rajoute une autre section dans le découpage des données,tout en gardant intacte nos données de test(20%),ce sont les données de **validation**. Pour comparer deux modèles de paramètres différents provenant d'un même modèle.

Ex: Deux KNeighborsClassifier avec un nombre de voisin différent.

On entraine les deux modèles sur les mêmes données et on sélectionne celui qui a la meilleur performance sur les données de validation avant le test final.

Cette section est une section intermédiaire entre celle de l'entrainement et celle du test. Elle sert a faire des paramétrages supplémentaires avant le test final et permet de palier au problème de surapprentissage.

##  cross-validation

Cet outil permet d'effectuer toutes les combinaisons possible quant au découpage des 80% des données en données de validation et d'entrainement.

# crossvalidation.png


````python
from sklearn.model_selection import cross_val_score
````

````python
cross_val_score(KNeighborsClassifier(2), X_train, Y_train, cv= 5, scoring='accuracy')
````

````python
cross_val_score(KNeighborsClassifier(2), X_train, Y_train, cv= 5, scoring='accuracy').mean()
````

````python
cross_val_score(KNeighborsClassifier(10), X_train, Y_train, cv= 5, scoring='accuracy')
````

````python
cross_val_score(KNeighborsClassifier(10), X_train, Y_train, cv= 5, scoring='accuracy').mean()
````

````python
val_score=[]
for k in range(1, 50):
  score = cross_val_score(KNeighborsClassifier(k), X_train, Y_train, cv=5).mean()
  val_score.append(score)
````

````python
plt.plot(val_score)
plt.grid(linestyle='--',which='both')
````

## GridSearchCv

Pour un modèle donné il peut y avoir plusieurs hyperparamètres, et différentes valeurs attribuées à ces paramètres peut ou ne pas produire un résultat optimal. Pour tester donc toutes les combinaisons possibles de tous les hyperparamètres d'un modèle on utilise ``GridSearchCv()``.

````python
from sklearn.model_selection import GridSearchCV
````

````python
param_grid = {'n_neighbors': np.arange(1, 20),
              'metric': ['euclidean','minkowski']}
              # on fait passer dans un dictionnaire le nom des hyperparamètres que l'on souhaite tester ainsi qu'un ensemble de valeurs
grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv= 5)
grid.fit(X_train, Y_train)
````

````python
grid.best_params_
````

````python
grid.best_score_
````

````python
model= grid.best_estimator_
````

````python
model.score(X_test, Y_test)
````

## Learning curve


GridSearchCv nous permet de trouver le modèle avec les meilleurs hyper paramètres en comparant les différentes performance de chaque combinaison grâce à la technique de crossvalidation.

Logiquement plus la machine dispose de données mieux devrait être sa performance. Cependant la performance atteint toujours un plafond au-delà duquel l'ajout de données serait inutile. Il est donc nécessaire de connaitre ce seuil, et cela est possible avec la fonction ``Learning curve`` qui permet de visualiser l'évolution de la courbe d'apprentissage en fonction d'une quantité de données croissant.

````python
from sklearn.model_selection import learning_curve
````

````python
N, train_score, val_score=learning_curve(model, X_train, Y_train, train_sizes=np.linspace(0.1, 1.0, 10), cv=5)
# le graphique représentera 10% jusqu'à 100% des données mentionné par le linspace
````

````python
print(N)
plt.plot(N, train_score.mean(axis=1), Label='train')
plt.plot(N, val_score.mean(axis=1), label = 'validation')
plt.xlabel('train_size')
plt.legend()
````

On observe qu'à un peu plus de 95% des données il y a une stagnation de la courbe d'apprentissage et celle de la validation.

````python
from sklearn.preprocessing import StandardScaler

iris = load_iris()
x = iris.data
outliers = np.full((10,4), 100)+np.random.randn(10, 4)
x = np.vstack((x,outliers))
x_minmax= MinMaxScaler().fit_transform(x)
x_stdscl = StandardScaler().fit_transform(x)

plt.scatter(x[:, 2], x[:, 3], alpha = 0.5, label='original' )
plt.scatter(x_minmax[:, 2], x_minmax[:, 3], alpha=0.2, label='minmax')
plt.scatter(x_stdscl[:, 2], x_stdscl[:, 3], alpha = 0.5, label='stand' )

plt.legend()

plt.show()
````

Nous remarquons que les données issues de la standardisation et du minmax sont très compressé donc difficiles à exploiter. Cela est dû à la présence des valeurs aberrantes auxquelles ces différents transformateurs sont sensibles.

### Robustscaler
Pour pallier à ce problème dû aux valeurs aberrantes ce transformateur est utilisé car moins sensible à eux.

````python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler

iris = load_iris()
x = iris.data

outliers = np.full((10,4), 100)+np.random.randn(10, 4)
x = np.vstack((x,outliers))

x_minmax= MinMaxScaler().fit_transform(x)
x_stdscl = StandardScaler().fit_transform(x)
x_robust = RobustScaler().fit_transform(x)


plt.scatter(x[:, 2], x[:, 3], alpha = 0.5, label='original' )
plt.scatter(x_minmax[:, 2], x_minmax[:, 3], alpha=0.2, label='minmax')
plt.scatter(x_stdscl[:, 2], x_stdscl[:, 3], alpha = 0.5, label='stand' )
plt.scatter(x_robust[:, 2], x_robust[:, 3], alpha = 0.5, label='robust' )

plt.legend()

plt.show()
````
